```python
import os
import pandas as pd

folder_path = "C:/Users/meets/nse-benny/eod2/src/eod2_data/daily"  
target_date = "2020-03-31"
close_prices = {}

for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)

        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        match = df[df['Date'] == target_date]

        file_key = file.replace(".csv", "").upper()
        if not match.empty:
            close_prices[file_key] = match.iloc[0]['Close']
        else:
            close_prices[file_key] = None

print(close_prices)

df = pd.DataFrame(list(close_prices.items()), columns=['STOCKNAME', 'PRICE0320'])
print(df.count())
df = df.dropna()
df.tail()
```

```python
import pandas as pd
import os

# Folder path where your CSV files are stored
folder_path = "C:/Users/meets/nse-benny/eod2/src/eod2_data/daily"

# Initialize an empty list to store individual DataFrames
all_data = []

# Loop through all CSV files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)
        
        # Check if both 'Date' and 'Close' columns exist
        if 'Date' in df.columns and 'Close' in df.columns:
            stock_name = os.path.splitext(filename)[0]
            # Extract and rename the columns
            temp_df = df[['Date', 'Close']].copy()
            temp_df.rename(columns={'Close': stock_name}, inplace=True)
            all_data.append(temp_df)

# Merge all DataFrames on the 'Date' column
from functools import reduce
combined_df = reduce(lambda left, right: pd.merge(left, right, on='Date', how='outer'), all_data)

# Convert 'Date' to datetime format for sorting (optional)
combined_df['Date'] = pd.to_datetime(combined_df['Date'])
combined_df.sort_values('Date', inplace=True)

# Reset index
combined_df.reset_index(drop=True, inplace=True)

# Show the result
print(combined_df.head())

```

```python
combined_df.to_csv("mega_dateclose.csv", index=False)
```

```python
import pandas as pd
import os

# Folder path where your CSV files are stored
folder_path = "C:/Users/meets/nse-benny/eod2/src/eod2_data/daily"

# Initialize an empty list to store individual DataFrames
all_data = []

# Loop through all CSV files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)
        
        # Check if both 'Date' and 'Close' columns exist
        if 'Date' in df.columns and 'Volume' in df.columns:
            stock_name = os.path.splitext(filename)[0]
            # Extract and rename the columns
            temp_df = df[['Date', 'Volume']].copy()
            temp_df.rename(columns={'Volume': stock_name}, inplace=True)
            all_data.append(temp_df)

# Merge all DataFrames on the 'Date' column
from functools import reduce
combined_df = reduce(lambda left, right: pd.merge(left, right, on='Date', how='outer'), all_data)

# Convert 'Date' to datetime format for sorting (optional)
combined_df['Date'] = pd.to_datetime(combined_df['Date'])
combined_df.sort_values('Date', inplace=True)

# Reset index
combined_df.reset_index(drop=True, inplace=True)

# Show the result
print(combined_df.head())

```

```python
combined_df.columns = [col.upper() if col.lower() != 'Date' else col for col in combined_df.columns]
print(combined_df.head())
```
