```python
import pandas as pd
```

```python
df1 = pd.read_excel("New_S3.xlsx")
df1.tail()
```

```python
#LOWER THE BETTER SO * -1 
mean_value = df1['Incremental CAPEX'].mean()
std_dev = df1['Incremental CAPEX'].std()
df1['ZSCORE_CAPEX'] = ((df1['Incremental CAPEX']-mean_value) / std_dev) * (-1)


#LOWER THE BETTER SO * -1
mean_value = df1['Incremental WC'].mean()
std_dev = df1['Incremental WC'].std()
df1['ZSCORE_WC'] = (df1['Incremental WC']-mean_value) / std_dev * (-1)


mean_value = df1['Sales Growth'].mean()
std_dev = df1['Sales Growth'].std()
df1['ZSCORE_SALES'] = (df1['Sales Growth']-mean_value) / std_dev


mean_value = df1['FH_Market Capitalization'].mean()
std_dev = df1['FH_Market Capitalization'].std()
df1['ZSCORE_MCAP'] = (df1['FH_Market Capitalization']-mean_value) / std_dev


df1.head()
```

```python
df2 = pd.read_excel("TRENDLYNE_230425.xlsx")
df2.head()
```

```python
df1['CD_BSE Code'] = pd.to_numeric(df1['CD_BSE Code'], errors='coerce')
df2['BSE Code'] = pd.to_numeric(df2['BSE Code'], errors='coerce')

df1['CD_NSE Symbol'] = df1['CD_NSE Symbol'].astype(str).str.upper()
df2['NSE Code'] = df2['NSE Code'].astype(str).str.upper()

# Deduplicate df2 to avoid one-to-many joins
df2_bse_dedup = df2.drop_duplicates(subset='BSE Code')
df2_nse_dedup = df2.drop_duplicates(subset='NSE Code')

# Step 1: Join on BSE Code
merged_bse = df1.merge(df2_bse_dedup, how='left', left_on='CD_BSE Code', right_on='BSE Code', indicator=True)

# Step 2: Filter unmatched and join them on NSE Code
unmatched = merged_bse[merged_bse['_merge'] == 'left_only'].copy()
matched = merged_bse[merged_bse['_merge'] == 'both'].copy()

# Remove merged df2 columns from unmatched before re-joining
unmatched_clean = unmatched[df1.columns]

# Join unmatched on NSE Code
merged_nse = unmatched_clean.merge(df2_nse_dedup, how='left', left_on='CD_NSE Symbol', right_on='NSE Code')

# Combine both
final_merged = pd.concat([matched.drop(columns=['_merge']), merged_nse], ignore_index=True)

# Optional: Save to Excel
# final_merged.to_csv("visual_tp.csv",index=False)
final_merged.info()
```

```python
# Set the DATE column as the index for rolling operation
column_list = final_merged['Stock Code'].tolist()

vol_df = pd.read_csv("C:/Users/meets/AVENTUS_FACTOR-MODEL/SIMPLE STRATEGY/mega_datevol_final.csv")
vol_df['DATE'] = pd.to_datetime(vol_df['DATE'])

vol_df.set_index('DATE', inplace=True)

# Filter column_list to only include columns present in vol_df
valid_columns = [col for col in column_list if col in vol_df.columns]

# Apply rolling mean with a window of 20 days and min_periods=1
rolling_means = vol_df[valid_columns].rolling(window=20, min_periods=1).mean()

# Extract the rolling average values as of 2009-04-01
rolling_avg_at_date = rolling_means.loc[pd.Timestamp('2025-03-03')]

# Store the result in a DataFrame
result_df_updated = pd.DataFrame({
    'STOCK NAME': rolling_avg_at_date.index,
    'MA': rolling_avg_at_date.values
})

# result_df_updated.info()

merged_df_1 = pd.merge(final_merged, result_df_updated, how='inner', left_on='Stock Code', right_on='STOCK NAME')
merged_df_1.info()

merged_df = merged_df_1[(merged_df_1["Market Capitalization"] >= 1000) & (merged_df_1["MA"] >= 10000)]

# Show the number of rows removed and preview the cleaned data
rows_removed = len(merged_df_1) - len(merged_df)
merged_df.head(), rows_removed
```

```python
df3 = pd.read_excel("C:/Users/meets/AVENTUS_FACTOR-MODEL/SIMPLE STRATEGY/L Valuation Data.xlsx")

final_merged_df = pd.merge(merged_df, df3, how='inner', left_on='Company Name', right_on='Company Name')
final_merged_df.head()

mean_value = final_merged_df['FR_Adjusted PE (x)'].mean()
std_dev = final_merged_df['FR_Adjusted PE (x)'].std()
final_merged_df['ZSCORE_PE'] = (final_merged_df['FR_Adjusted PE (x)']-mean_value) / std_dev

mean_value = final_merged_df['FR_EV/EBITDA(x)'].mean()
std_dev = final_merged_df['FR_EV/EBITDA(x)'].std()
final_merged_df['ZSCORE_EVEBITDA'] = (final_merged_df['FR_EV/EBITDA(x)']-mean_value) / std_dev

final_merged_df['ZSCORE_FINAL'] = ((1/5) * final_merged_df['ZSCORE_CAPEX']) + ((1/5) * final_merged_df['ZSCORE_WC']) + ((1/5) * final_merged_df['ZSCORE_SALES']) + (0.20 * final_merged_df['ZSCORE_MCAP']) + (0.10 * final_merged_df['ZSCORE_PE']) + (0.10 * final_merged_df['ZSCORE_EVEBITDA'])
# final_merged_df['ZSCORE_FINAL'] = (1 * final_merged_df['ZSCORE_CAPEX'])



final_merged_df = final_merged_df[final_merged_df["ZSCORE_FINAL"]>0]

# final_merged_df['PORTFOLIO_WEIGHTS'] = final_merged_df["ZSCORE_FINAL"]/final_merged_df["ZSCORE_FINAL"].sum()

column_list = final_merged_df['Stock Code'].tolist()
print(column_list)

final_merged_df.head()

final_merged_df.to_csv("final_s32025.csv")
```

```python
df = pd.read_csv("C:/Users/meets/AVENTUS_FACTOR-MODEL/SIMPLE STRATEGY/mega_dateclose_final.csv")
columns_to_keep = ['Date'] + [col for col in column_list if col in df.columns]
filtered_with_date_df = df[columns_to_keep]


filtered_with_date_df['Date'] = pd.to_datetime(filtered_with_date_df['Date'])
start_date = pd.to_datetime("2018-03-31")
end_date = pd.to_datetime("2021-04-02")
filtered_df = filtered_with_date_df[(filtered_with_date_df['Date'] >= start_date) & (filtered_with_date_df['Date'] <= end_date)]
filtered_df.head()


df_cleaned = filtered_df.dropna(how='all', subset=filtered_df.columns.difference(['Date']))
rows_removed = len(filtered_df) - len(df_cleaned)
# print("ROWS REMOVED i.e. ALL VALUES MISSING EXCEPT DATE:", rows_removed)
df_cleaned.head()


df_fully_cleaned = df_cleaned.dropna(axis=1)
final_columns = df_fully_cleaned.columns.tolist()
if 'Date' in final_columns:
    final_columns.remove('Date')
print(final_columns)
df_fully_cleaned.head()


df_fully_cleaned['Date'] = pd.to_datetime(df_fully_cleaned['Date'])
df_fully_cleaned.set_index('Date', inplace=True)
pct_change_df = df_fully_cleaned.pct_change() 
pct_change_df = pct_change_df.iloc[1:] 
pct_change_df.head()
# pct_change_df.to_csv("gptwork.csv",index=False)



pct_change_df = pct_change_df.reset_index()  # This moves 'Date' from index to a regular column
pct_change_df.head()
# pct_change_df.to_csv("data1.csv",index=False)





# pct_change_df.to_csv("deletenow.csv",index=False)
# nan_counts = pct_change_df.isna().sum()
# print(nan_counts)
# columns_without_nans = df.columns[df.isna().sum() == 0].tolist()
# print(columns_without_nans)

# pct_change_df.dropna
```

```python
# PORTFOLIO WT CALCULATION

final_df = final_merged_df[final_merged_df['Stock Code'].isin([name.upper() for name in final_columns])]
final_df.head()

final_df['PORTFOLIO_WEIGHTS'] = final_df["ZSCORE_FINAL"]/final_df["ZSCORE_FINAL"].sum()
final_df.head()

final_df = final_df[["Stock Code","PORTFOLIO_WEIGHTS"]]
final_df.head(20)
# final_df.to_csv("data2.csv",index=False)
```

```python
#DIRECT COPYPASTE - NO EDIT


# Align columns: ensure weights match the columns in the return dataframe
returns_df = pct_change_df.copy()
weights_df = final_df.set_index("Stock Code")["PORTFOLIO_WEIGHTS"]

# Ensure only common stocks are used
common_stocks = returns_df.columns.intersection(weights_df.index)
returns_matrix = returns_df[common_stocks].copy()
weights_vector = weights_df.loc[common_stocks]

# Perform matrix multiplication (dot product) for each date
returns_df["Portfolio Return"] = returns_matrix.dot(weights_vector)

# Show a few results
final_portfolio = returns_df[["Date", "Portfolio Return"]]
final_portfolio.info()
```
