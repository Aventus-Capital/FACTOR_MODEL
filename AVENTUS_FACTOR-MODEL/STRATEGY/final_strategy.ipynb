```python
import pandas as pd
```

```python
df1 = pd.read_excel("CAPEX L15.xlsx")
df1 = df1.iloc[:-3]
df1.tail()
```

```python
df1["GB+CWIP_CAGR_SUM"] = df1["GB+CWIP_1Y_CAGR"] + df1["GB+CWIP_2Y_CAGR"] + df1["GB+CWIP_3y_CAGR"]

print("mean std for cagr")
mean_value = df1['GB+CWIP_CAGR_SUM'].mean()
print(mean_value)
std_dev = df1['GB+CWIP_CAGR_SUM'].std()
print(std_dev)

df1['ZSCORE_CAGR'] = (df1['GB+CWIP_CAGR_SUM']-mean_value) / std_dev

print("mean std for mcap")
mean_value = df1['FH_Market Capitalization'].mean()
print(mean_value)
std_dev = df1['FH_Market Capitalization'].std()
print(std_dev)

df1['ZSCORE_MCAP'] = (df1['FH_Market Capitalization']-mean_value) / std_dev

df1.tail()
```

```python
import os
import pandas as pd

folder_path = "C:/Users/meets/nse-benny/eod2/src/eod2_data/daily"  
target_date = "2009-03-31"
close_prices = {}

for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)

        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        match = df[df['Date'] == target_date]

        file_key = file.replace(".csv", "").upper()
        if not match.empty:
            close_prices[file_key] = match.iloc[0]['Close']
        else:
            close_prices[file_key] = None

print(close_prices)

df = pd.DataFrame(list(close_prices.items()), columns=['STOCKNAME', 'PRICE0309'])
print(df.count())
df = df.dropna()
df.tail()
```
